```python
# Describe the number of classes:
num_class = 9

from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Conv2DTranspose, Dense, Activation, Lambda, Reshape, Flatten,Dropout,BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping

# Custom classifier function:
def classifier_func(x):
    return x+x*K.one_hot(K.argmax(x, axis=1), num_classes=num_class)
```

## Please build the complex architecture with droupout and L2 regularizations

### Version 1

You can use this formula [(W−K+2P)/S]+1.

    - W is the input volume
    - K is the Kernel size 
    - P is the padding 
    - S is the stride.

```python
from tensorflow.keras.optimizers import Adam
# Deep Learning Model: (basic model)

inputs = Input(shape=(28, 28, 3))
#Encoder:
conv_1 = Conv2D(32, (3,3), strides=(1,1))(inputs)

act_1 = Activation('relu')(conv_1)

maxpool_1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(act_1)

conv_2 = Conv2D(64, (3,3), strides=(1,1), activity_regularizer=regularizers.l2(10e-5), padding='same')(maxpool_1)

act_2 = Activation('relu')(conv_2)

maxpool_2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(act_2)

# Output Shape: 6x6x64
    
flat_1 = Flatten()(maxpool_2)
drop1=Dropout(0.2)(flat_1)

fc_1 = Dense(256)(drop1)
act_3 = Activation('relu')(fc_1)

fc_2 = Dense(128)(act_3)
act_4 = Activation('relu')(fc_2)

fc_3 = Dense(num_class)(act_4)

act_class = Lambda(classifier_func, output_shape=(num_class,))(fc_3)
# Output Shape: 9

#Decoder:
fc_4 = Dense(256)(act_class)
act_5 = Activation('relu')(fc_4)

fc_5 = Dense(2304)(act_5)
act_6 = Activation('relu')(fc_5)
reshape_1 = Reshape((6,6,64))(act_6)

upsample_1 = UpSampling2D((2, 2))(reshape_1)
deconv_1 = Conv2DTranspose(64, (3, 3), strides=(1, 1))(upsample_1)
act_7 = Activation('relu')(deconv_1)

upsample_2 = UpSampling2D((2, 2))(act_7)
deconv_2 = Conv2DTranspose(32, (3, 3), strides=(1, 1), activity_regularizer=regularizers.l2(10e-5))(upsample_2)
act_8 = Activation('relu')(deconv_2)

conv_3 = Conv2D(3, (3, 3), strides=(1, 1))(act_8)
act_9 = Activation('sigmoid')(conv_3)
# Output Shape: 28x28x1

autoencoder = Model(inputs, act_9)

#autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
autoencoder_optimizer = tensorflow.keras.optimizers.Adam(learning_rate=0.0004) 

# We use binary crossentropy loss here. You can try Wasserstein loss to see the difference.
autoencoder.compile(optimizer=autoencoder_optimizer, loss='mse',metrics=['acc'])
autoencoder.summary()
```

```python
# Fit the model and record the training history for plotting purpose
history=autoencoder.fit(train_gen,
                        batch_size=64,
                        steps_per_epoch=20,
                        epochs=20, 
                        validation_data=val_gen)
```

```python
autoencoder.evaluate(test_gen, verbose=1)
```

```python
# Plot history
import matplotlib.pyplot as plt
plt.plot(history.history['acc'], color='b')
plt.plot(history.history['val_acc'], color='g')
plt.show()
```

```python
# Deep Learning Model: add early stopping and binary cross entropy loss function

inputs = Input(shape=(28, 28, 3))
#Encoder:
conv_1 = Conv2D(32, (3,3), strides=(1,1))(inputs)
act_1 = Activation('relu')(conv_1)
maxpool_1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(act_1)

conv_2 = Conv2D(64, (3,3), strides=(1,1), padding='same')(maxpool_1)
act_2 = Activation('relu')(conv_2)
maxpool_2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(act_2)
# Output Shape: 6x6x64
    
flat_1 = Flatten()(maxpool_2)
drop1=Dropout(0.2)(flat_1)
fc_1 = Dense(256)(drop1)
act_3 = Activation('relu')(fc_1)

fc_2 = Dense(128)(act_3)
act_4 = Activation('relu')(fc_2)

fc_3 = Dense(num_class)(act_4)

act_class = Lambda(classifier_func, output_shape=(num_class,))(fc_3)
# Output Shape: 10

#Decoder:
fc_4 = Dense(256)(act_class)
act_5 = Activation('relu')(fc_4)

fc_5 = Dense(2304)(act_5)
act_6 = Activation('relu')(fc_5)
reshape_1 = Reshape((6,6,64))(act_6)

upsample_1 = UpSampling2D((2, 2))(reshape_1)
deconv_1 = Conv2DTranspose(64, (3, 3), strides=(1, 1))(upsample_1)
act_7 = Activation('relu')(deconv_1)

upsample_2 = UpSampling2D((2, 2))(act_7)
deconv_2 = Conv2DTranspose(32, (3, 3), strides=(1, 1))(upsample_2)
act_8 = Activation('relu')(deconv_2)

conv_3 = Conv2D(3, (3, 3), strides=(1, 1))(act_8)
act_9 = Activation('sigmoid')(conv_3)
# Output Shape: 28x28x1

autoencoder = Model(inputs, act_9)
#es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=5)
#autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
autoencoder_optimizer = keras.optimizers.Adam(learning_rate=0.0004, clipvalue=1.0, decay=1e-8, momentum=0.9) #for avoid exploiding gradient, we use gradient clipping 
# We use binary crossentropy loss here. You can try Wasserstein loss to see the difference.
autoencoder.compile(optimizer=autoencoder_optimizer, loss='binary_crossentropy',metrics=['acc'])
autoencoder.summary()
```
